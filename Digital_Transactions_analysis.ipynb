# Feature Engineering

```python
def category(amount):
    if amount < 1000:
        return "Small"
    elif amount < 5000:
        return "Medium"
    else:
        return "Large"

df["transaction_size"] = df["amount (INR)"].apply(category)
```

```python
df["day_type"] = df["is_weekend"].map(
    {
        1: "Weekend",
        0: "Weekday"
    }
)
```

```python
df["status_code"] = df["transaction_status"].map(
    {
        "SUCCESS": 1,
        "FAILED": 0
    }
)
```
